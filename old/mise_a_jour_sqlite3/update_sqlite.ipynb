#### hyperion 1 --> DONE

```python
# hyperion :

import sqlite3

# Chemin vers SQLite
database_path = 'new file/statistics.sqlite3'

# Connexion à la base de données SQLite
conn = sqlite3.connect(database_path)
cursor = conn.cursor()

# Récupérer les valeurs agrégées pour 'total_read_time' et 'total_read_pages'
query = """
    SELECT 
        SUM(total_read_time) AS total_read_time,
        SUM(total_read_pages) AS total_read_pages
    FROM book
    WHERE id IN (15, 12, 20);
"""
cursor.execute(query)
total_read_time, total_read_pages = cursor.fetchone()

# Mettre à jour l'entrée avec id = 20 en conservant les informations voulues
update_query = f"""
    UPDATE book
    SET 
        total_read_time = {total_read_time},
        total_read_pages = {total_read_pages}
    WHERE id = 20;
"""
cursor.execute(update_query)

# Supprimer les lignes avec id 15 et 12 maintenant agrégées
delete_query = "DELETE FROM book WHERE id IN (15, 12);"
cursor.execute(delete_query)

# Valider les changements
conn.commit()

# Fermer la connexion
conn.close()


```

remplacer les id sur la colonne id_book

```python

# Chemin vers SQLite
database_path = 'new file/statistics.sqlite3'

# Connexion à la base de données SQLite
conn = sqlite3.connect(database_path)
cursor = conn.cursor()

# Remplacements à effectuer
remplacements = {
    15: 20,
    12: 20,
    22: 23,
    17: 4,
    14: 25,
    10: 11
}

# Appliquer les modifications dans SQLite
for old_value, new_value in remplacements.items():
    query = f"UPDATE page_stat_data SET id_book = {new_value} WHERE id_book = {old_value};"
    cursor.execute(query)

# Valider les changements
conn.commit()

# Fermer la connexion
conn.close()
```
